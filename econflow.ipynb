```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function


import tensorflow as tf
import numpy as np

FLAGS = None



population = 4

h = tf.Variable(tf.ones([population], tf.float32))

w = tf.constant(np.exp(np.random.rand(population).astype(np.float32))) ## TODO: generalize
tax_rate = tf.Variable(0.1)

sess = tf.InteractiveSession()
```

```python
sess.run(tf.global_variables_initializer())
print(sess.run(w))


## take the maximum so that you can't work negative hours
pre_tax_incomes = tf.maximum(tf.multiply(h, w), 0)

actual_tax_rate = tf.maximum(tf.minimum(1.0, tax_rate), 0.0)
total_income = tf.reduce_sum(tf.multiply(h, w))
total_tax_revenue = total_income * tax_rate
ubi = total_tax_revenue / population
post_tax_incomes = pre_tax_incomes * (1 - tax_rate) + ubi

utilities = tf.log(post_tax_incomes) - h * 2
[d_utilities_on_d_h] = tf.gradients(utilities, [h])

social_welfare = tf.reduce_sum(utilities)
d_social_welfare_on_d_tax_rate = tf.gradients(social_welfare, [tax_rate])[0]

entropy = tf.reduce_sum(tf.multiply(d_utilities_on_d_h, d_utilities_on_d_h)) + d_social_welfare_on_d_tax_rate ** 2

train_step = tf.train.GradientDescentOptimizer(0.01).minimize(entropy)

for _ in range(100):
    print(sess.run([h, tax_rate, ubi])) # , post_tax_incomes, tax_rate, social_welfare, ubi
    sess.run(train_step)
```
